```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
def initialize_with_zeros(layer_dims):
    params = {}
    L = len(layer_dims)
    for l in range(1, L):
        params['W' + str(l)] = np.zeros((layer_dims[l], layer_dims[l - 1]))
        params['b' + str(l)] = np.zeros((layer_dims[l], 1))

    return params
```

```python
def intialize_random(layer_dims):
    params = {}
    L = len(layer_dims)
    for l in range(1, L):
        params['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l - 1]) * 10
        params['b' + str(l)] = np.zeros((layer_dims[l], 1))
    
    return params
```

```python
def He_initialization(layer_dims):
    params = {}
    L = len(layer_dims)
    for l in range(1, L):
        params['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l - 1]) * (np.sqrt(2/ layer_dims[l - 1]))
        params['b' + str(l)] = np.zeros((layer_dims[l], 1))

    return params
```
